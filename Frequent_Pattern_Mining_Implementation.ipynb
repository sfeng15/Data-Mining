```python
'''
    step 1,2
'''
import re
import operator
vocab = open("vocab.txt", "w")
title = open("title.txt", "w")

'''
    build the dictionay, key:word, value:ID for that word
    uniqueNum increment by one each iteration
'''
wordDic={} #global dictionary
uniqueNum=0
for index, line in enumerate(open('paper.txt', 'r').readlines()):

    w=re.split('\t| |\n',line)
    '''
        cut the id, newline, use the rest title as key
    '''
    del w[-1]
    del w[0]
    for i in range(len(w)):
        if wordDic.has_key(w[i]):
            continue
#         print w[i]
#         print uniqueNum
        wordDic[w[i]]=uniqueNum
        uniqueNum= uniqueNum + 1     
'''
    write the dictionary to the output file
'''
sorted_dic=sorted(wordDic.items(),key=operator.itemgetter(1))
for word,id in sorted_dic:
    vocab.write("%s\n" % word)
# print wordDic
    
    
'''
    build the text by dictionary    
'''
dic={}#local dic for each title
for index, line in enumerate(open('paper.txt', 'r').readlines()):
    
    '''
    cut the id and end line
    '''
    w=re.split('\t| |\n',line)   
    del w[-1]
    del w[0]
    
    '''
    build the local dic for each line
    '''
    for i in range(len(w)):
        if dic.has_key(w[i]):
            dic[w[i]]=dic[w[i]]+1
        else:
            dic[w[i]]=1
#     print w
#     print dic
    
    '''
    build the new string with total word occurance(len(dic)),word:occurance
    '''
    res=""
    res=res+str(len(dic))+" "
    s=set()
    for i in range(len(w)):
#         print w[i]
#         print wordDic[w[i]]
        if w[i] in s:
            continue
        s.add(w[i])
        res=res+str(wordDic[w[i]])+":"+str(dic[w[i]])+" "
    res=res+"\n"
    
    title.write(res)
    '''
    clear the local dictionary for next line
    '''
    dic.clear()
    
    
vocab.close()
title.close()
```
